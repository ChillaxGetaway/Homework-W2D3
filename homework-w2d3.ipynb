```python
ShoppingCart=[]
ProdPrice=[]
session="y"
confirm="y"

class Operations:
    def __init__(self):
        print("Welcome!")
        print("Thanks for starting your shopping session!")

    def add(self):
        confirm="y"
        while confirm=="y":    
            product=input("Type the product name...\n").title()
            ShoppingCart.append(product)
            price=input("Type the price for "+product+"...\n")
            if price.isnumeric :
                ProdPrice.append(int(price))
            else:
                print("maybe")
            
            confirm=input("Type 'y' to keep adding products or type 'n' to continue you shopping session.\n").lower()
            if confirm=="y":
                pass
            elif confirm=="n":
                pass
            else:
                print("That is not a valid selection!")
                confirm="n"

    def delete(self):
        confirm="y"
        while confirm=="y":
            selection=input("What product you want to remove?\n").title()
            confirm=input("Are you sure you want to delete "+selection+".\nType 'y' to confirm or type 'n' to deny and continue you shopping session.\n").lower()
            if confirm=="y":
                if selection not in ShoppingCart:
                    print("That is not a product you have in your cart...")
                else:
                    ProdPrice.pop(ShoppingCart.index(selection))
                    ShoppingCart.remove(selection)
                    print(selection+" has been deleted successfully!\n")
            elif confirm=="n":
                print("Ok, "+ selection+"will not be removed from your cart.\n")
            else:
                print("That is not a valid selection!")
                confirm="n"

            confirm=input("Type 'y' to keep deleting products or type 'n' to continue you shopping session.\n").lower()
            if confirm=="y":
                pass
            elif confirm=="n":
                pass
            else:
                print("That is not a valid selection!")
                confirm="n"

    def view(self):
        print(ShoppingCart)
        print(ProdPrice)

    def quit(self,):
        confirm=input("Are you sure you want to finish your shopping session?\nType 'y' to confirm or type 'n' to deny and continue you shopping session.\n").lower()
        if confirm=="y":
            if len(ShoppingCart):
                print(f'Your shopping cart has these products: {ShoppingCart}')
                print('The Total price of your shopping cart is: $', sum(ProdPrice))
                print("Thanks for shopping with us!!!")
                session="n"
            else:
                print("Your cart is empty")
                print("Thanks for shopping with us!!!")
                session="n"
        elif confirm=="n":
            pass
        else:
            print("That is not a valid selection!")
            pass

my_opp = Operations() 

while session=="y":
    action=input("Type the action you want for your Shopping cart. \n Add | Delete | View | Quit \n").lower()
    if action == 'add':
        my_opp.add() # in order to access the functions you created, you need to
    elif action == 'delete':
        my_opp.delete()
    elif action ==  'view':
        my_opp.view()
    elif action == 'quit':
        my_opp.quit()
        session="n"
    else:
        print("C'mon  yo...look at the keyboard when you type ._.")



```
